```python
# Import the SchemaMapEngine class

from metaharmonizer.models.schema_mapper import SchemaMapEngine
```

```python
# Choose the clinical data file

file = "data/ucec_cptac_2020_before_harmonization.csv"
```

```python
# Initialize the SchemaMapEngine
engine = SchemaMapEngine(file, mode="auto", top_k=5)


# Run the schema mapping
results = engine.run_schema_mapping()

# Persist the auto-mode predictions to a stable path for the evaluation
# cell below. run_schema_mapping() also writes a timestamped copy under
# SM_OUTPUT_DIR; this fixed filename is what the eval cell reads.
results.to_csv("data/ucec_cptac_2020_auto.csv", index=False)
```

```python
# Preview the top of the schema-mapping results
results.head()
```

```python
# Option: evaluation

# Import the schema mapping evaluation module
from functions import schema_mapping_evaluation

# Ground truth (data/sm_truth.csv) is a curated cBioPortal "source,ref_match"
# mapping of raw column names -> standard fields. Only the input columns that
# appear as a source in this mapping are scored (the UCEC clinical columns:
# AGE, SEX, RACE, ETHNICITY, COUNTRY, CANCER_TYPE, CANCER_TYPE_DETAILED, TUMOR_SITE).
metrics = schema_mapping_evaluation.compute_accuracy(
    pred_file="data/ucec_cptac_2020_auto.csv",
    truth_file="data/sm_truth.csv",
    top_k=5,
    #exclude_methods=("alias",),
    #include_methods=("value",),
    save_eval=True,
    out_dir="data"
)
print(metrics)
```
